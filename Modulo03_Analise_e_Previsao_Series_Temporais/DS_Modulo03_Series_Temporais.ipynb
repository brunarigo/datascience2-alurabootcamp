#Modulo 03 - Análise e Previsão de Séries Temporais

## Aula 01 - Entendendo Séries Temporais

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
```

```python
url = 'https://github.com/allanspadini/Prophet_covid/raw/main/dados_covid_sp.zip'
```

```python
dados = pd.read_csv(url, sep=";")
```

```python
dados.head()
```

```python
dados['datahora'] = pd.to_datetime(dados['datahora'], format='%Y-%m-%d')
```

```python
import matplotlib as mpl
```

```python
mpl.rcParams['font.size'] = 18
mpl.rcParams['figure.figsize'] = (20,8)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x='datahora', y='casos',data=dados)
```

```python
dados_sp = dados.query('nome_munic == "São Paulo"')
```

```python
dados_sp.head()
```

```python
sns.lineplot(x='datahora',y='casos',data=dados_sp)
```

```python
import numpy as np
```

```python
#Modelo de grafico generico exponencial
eixo_x = np.linspace(-2, 2, 100)
eixo_y = np.exp(eixo_x)
sns.lineplot(x=eixo_x,y=eixo_y)
```

```python
#Colocando em escala de LOG
sns.lineplot(x=eixo_x,y=eixo_y)
mpl.pyplot.yscale('log')
```

```python
sns.lineplot(x='datahora',y='casos',data=dados_sp)
mpl.pyplot.yscale('log')
```
