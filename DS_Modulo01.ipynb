###Desafio 01: Escolher Título mais descritivo, que passe a mensagem adequada.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação",y="2008/Ago", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor gasto em Internações no SUS por UF período de Agosto/2008")
plt.show()
```

###Desafio 02: Realizar a análise anterior para mês mais recente

```python
ax = dados.plot(x="Unidade da Federação",y="2021/Mar", kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor gasto em Internações no SUS por UF período de Março/2021")
plt.show()
```
