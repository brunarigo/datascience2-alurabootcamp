##Aula 2 - Primeiras visualizações de Dados

**Estudos realizados utilizando os dados provenientes da Data de Atendimento**


```python
dados = pd.read_csv("/content/drive/MyDrive/Bootcamp Data Science/Modulo 01/Arquivos_Modulo01/A091915189_28_143_208_DTAtendimento.csv", encoding="ISO-8859-1", engine="python",
                    skiprows = 3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")
dados.head()
```

```python
dados[['2008/Ago','2008/Set']].head()
```

```python
dados.mean()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index("Unidade da Federação")
usaveis
```

```python
usaveis["2019/Ago"]
```

```python
usaveis.loc["12 Acre"]
```

```python
usaveis.T.head()
```

```python
usaveis.T.plot(figsize=(10,6))
```

```python
usaveis = usaveis.drop("Total", axis=1)
usaveis.head()
```

```python
import matplotlib.pyplot as plt

usaveis.T.plot(figsize=(10,6))
plt.show()
```

###Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada
Também foram ajustados os valores e casas decimais.

```python
import matplotlib.ticker as ticker

ax = usaveis.T.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

###Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
ax = top5.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

###Desafio 2: Adicionar uma coluna mostrando a Região de cada Estado.

```python
#usaveis["Região"] = ["Região Norte", "Região Norte", "Região Norte", "Região Norte", "Região Norte", "Região Norte", "Região Norte",
#                     "Região Nordeste", "Região Nordeste", "Região Nordeste", "Região Nordeste", "Região Nordeste", "Região Nordeste", "Região Nordeste", "Região Nordeste", "Região Nordeste",
#                     "Região Sudeste", "Região Sudeste", "Região Sudeste", "Região Sudeste",
#                     "Região Sul", "Região Sul", "Região Sul",
#                     "Região Centro-Oeste", "Região Centro-Oeste", "Região Centro-Oeste", "Região Centro-Oeste"]
```

```python
usaveis.head()
```

```python
#Ordenar pela colula Total (menor para maior por padrão)
ordenado_por_total = usaveis.sort_values("Total", ascending=False)
ordenado_por_total = ordenado_por_total.drop("Total", axis=1)
ordenado_por_total.head()
```

###Desafio 3: Deixar a mensagem mais clara (titulo, legenda, etc).

```python
#Apresenta os 5 primeiros que mais gastaram
ax = ordenado_por_total.head(5).T.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.xlabel("Mês/Ano")
plt.ylabel("Valor")
plt.title("UFs com maiores gastos de Internações no SUS")
plt.show()
```

###Desafio 4: Adicione um estado de preferência à lista dos 5 estados que tiveram maior gasto

```python
#Exclui as primeiras colunas de valores discrepantes - seleciona somente a partir de 6, neste caso
colunas_interessadas = ordenado_por_total.columns[6:]
ordenado_por_total = ordenado_por_total[colunas_interessadas]
```

```python
#Adiciona as UFs de preferência
estados = ordenado_por_total.loc [['35 São Paulo', '31 Minas Gerais','41 Paraná','43 Rio Grande do Sul','33 Rio de Janeiro','42 Santa Catarina']]
ax = estados.T.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.xlabel("Mês/Ano")
plt.ylabel("Valor")
plt.title("UFs escolhidas e seus gastos com Internações no SUS")
plt.show()
```

###Desafio 5: Plotar somente região Sudeste e ver se afetou com um pico de forma similar
